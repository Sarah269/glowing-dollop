## National Gallery of Art Web Scrape

```python
# import libraries
from bs4 import BeautifulSoup
import requests
import smtplib
import time
import datetime

```

### Read web page

```python
page = requests.get('https://web.archive.org/web/20161025170331/http://www.nga.gov/collection/anA1.htm')

soup = BeautifulSoup(page.text, 'html.parser')
```

```python
## Parse html to obtain values
```

```python
# Pull all text from the BodyText div
artist_name_list = soup.find(class_='BodyText')

# Pull text from all instances of <a> tag within BodyText div
artist_name_list_items = artist_name_list.find_all('a')

# Create for loop to print out all artists' names
for artist_name in artist_name_list_items:
    print(artist_name.prettify())
```

```python
# Remove bottom links
last_links = soup.find(class_='AlphaNav')
last_links.decompose()

artist_name_list = soup.find(class_='BodyText')
artist_name_list_items = artist_name_list.find_all('a')

for artist_name in artist_name_list_items:
    print(artist_name.prettify())
```

```python
# Initialize 
Artist_nm = []
Artist_lnk = []

# Read artist name and get link reference
# Append names and links to Series
# Print names and links
for artist_name in artist_name_list_items:
    names = artist_name.contents[0]
    links = 'https://web.archive.org' + artist_name.get('href')
    Artist_nm.append(names)
    Artist_lnk.append(links)
    print(names)
    print(links)
```

### Create Dataframe

```python
import pandas as pd
# set max column width 
pd.set_option('display.max_colwidth', None)

# Create dataframe
Gallery_artists = pd.DataFrame({'Name':Artist_nm, 'Link':Artist_lnk})
Gallery_artists
```

```python
#  Create csv file
Gallery_artists.to_csv('Gallery_Artists.csv',index=False, encoding='UTF-8')
```
